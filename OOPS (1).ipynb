# **INHERITANCE**


Inheritance allows us to defines a class that inherits all the methods and properties from another class.Parent class is the class being inherited from, also called base class.Child class is the class that inherited from another class,also called derived class.

Reuse of code.

Add extra functionality to existing class

parent = base class

SINGLE INHERITANCE


Single inheritance allows to derivate a class to inherit properties of one parent class.

```python
#single inheritance

class A:

  def printa(self,num):
    self.num = num
    print('inside class A')

class B(A):
  def printb(self):
    print('inside class B')

obj = B()
obj.printa(3)
obj.printb()
```

```python
class Person:
  def __init__(self,name):
    name = input('Enter your name : ')
    self.name = name

  def new(self):
    print(self.name)

class Person2(Person):
  def __init__(self,name,age):
    Person.__init__(self,name)
    age = int(input('Enter your age : '))
    self.age = age
  def new1(self):
    print(self.name)
    print(self.age)


obj = Person2('shaluuu',20)
obj.new1()
```

**MULTIPLE INHERITANCE**


If a classis able to be createdfrom multiple base classes this kind of inheritance is known as multiple inheritance.

When multiple inheritance, each of the attributes that present in the classes of the base has been passed on to the class that is derived from it.

```python
class A:
  def printa(self):
    print('inside class A')


class B:

  def printb(self):
    print('inside class B')

class C(A,B):
  def printc(self):
    print('inside class C')

obj = C()
obj.printa()
obj.printb()
obj.printc()
```

**MULTI LEVEL INHERITANCE**


The features that are part of the original class, as well as the class that is derived from it, are passed on to the new class. It is similar to the relationship involving grantparents and children.

```python
class A:
  def printa(self):
    print('inside class A')

class B(A):
  def printb(self):
    print('inside class B')

class C(B):
  def printc(self):
    print('inside class C')


obj = C()
obj.printa
obj.printb
obj.printc
```
